```python
from langchain_community.retrievers import BM25Retriever

retriever = BM25Retriever.from_texts(["Australia won the Cricket World Cup 2023",
 "India and Australia played in the finals",
 "Australia won the sixth time having last won in 2015"])

query = "Who won the 2023 Cricket World Cup?"
docs = retriever.invoke(query)
print(f"\nFound {len(docs)} documents similar to the query: '{query}'")
for i, doc in enumerate(docs):
    print(f"\nDocument {i + 1} (first 200 characters):")
    print(doc.page_content[:200])
```
